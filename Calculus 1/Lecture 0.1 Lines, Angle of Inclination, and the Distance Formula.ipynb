# Calculus 1 Lecture 0.1: Lines, Angle of Inclination, and the Distance Formula

A Line is defined as a figure that does not have a beginning or an end, contains infinite points and does not bend. A general form of the line is shown below.

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import math
import random
```

```python
point_A = (3, 1)
point_B = (9, 4)

fig, ax = plt.subplots()
plt.xticks([point_A[0], point_B[0]])
plt.yticks([point_A[1], point_B[1]])

def draw_line(ax, x1, y1, x2, y2):
    plt.plot([x1, x2], [y1, y2], color="blue")
    plt.scatter([x1, x2], [y1, y2], color="red")

    ax.set_xlim([0, 10])
    ax.set_ylim([0, 10])

    
draw_line(ax, *point_A, *point_B)
```

In order to mathematically define a line, we need two parameters: **slope** and **intercept**, which are usually denoted as $m$ and $b$, making the mathematical definition of the line as follows:

\begin{align*}
y = mx + b
\end{align*}

To define the slope, one can take `point_A` and `point_B` defined in the code, and apply the following formula:

\begin{align*}
m = \frac{rise}{run} = \frac{{y_{B} - y_{A}}}{{x_{B} - x_{A}}}
\end{align*}

```python
def find_slope(x1, y1, x2, y2):
    rise = y2 - y1
    run = x2 - x1
    slope = rise / run
    return slope

print("The slope of the line is:", find_slope(*point_A, *point_B))
```

Multiplying both sides of the equation for $m$ by $x_{B}-x_{A}$ gives us the **point-slope** equation, where the line is defined by using its slope and a point that it passes through:

\begin{align*}
m(x_{B} - x_{A}) = y_{B} - y_{A}
\end{align*}

Let us assume that $x_{B}=x$ and $y_{B}=y$. This simplifies the equation to the following form:

\begin{align*}
mx - mx_{A} = y - y_{A}
\end{align*}

Since we are given that the line passes through the `point_A`, which is ($x_{A}$, $y_{A}$), we can rearrange the equation into the following form:

\begin{align*}
y = mx + y_{A} - mx_{A} = mx + (y_{A} - mx_{A})
\end{align*}

Here, the expression $(mx_{A} - y_{A})$ consists of three known values, which means that they can be replaced by a simple $b$ and give us the **slope-intercept** equation:

\begin{align*}
y = mx + b
\end{align*}

The term $b$ is called **intercept**, since it defines how the line intercepts with the $y$ axis when $x=0$.

We can also have constant lines, such as $y = c$ or $x = c$, where $c$ is a constant.

```python
c = random.randint(1, 5)

fig, ax = plt.subplots()
plt.axhline(y=c, color="r", linestyle="--", label=f"y = {c}")
plt.axvline(x=c, color="b", linestyle=":", label=f"x = {c}")

ax.set_xlim(c - 5, c + 5)
ax.set_ylim(c - 5, c + 5)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_title("Horizontal and Vertical Lines")

ax.legend()
```

### Parallel Lines: Lines that have the same slope
### Perpendicular Lines: Lines that have the slopes which are the negative reciprocals of each-other

```python
fig, ax = plt.subplots()
fig.suptitle("Parallel Lines")

parallel_point1 = (1, 1)
parallel_point2 = (5, 6)
parallel_point3 = (2, 1)
parallel_point4 = (8, 8.5)

plt.xticks([parallel_point1[0], parallel_point2[0], parallel_point3[0], parallel_point4[0]])
plt.yticks([parallel_point1[1], parallel_point2[1], parallel_point3[1], parallel_point4[1]])

draw_line(ax, *parallel_point1, *parallel_point2)
draw_line(ax, *parallel_point3, *parallel_point4)

print(f"Slope of the first line:\t{find_slope(*parallel_point1, *parallel_point2)}")
print(f"Slope of the second line:\t{find_slope(*parallel_point3, *parallel_point4)}")
```

```python
fig, ax = plt.subplots()
fig.suptitle("Perpendicular Lines")

perpendicular_point1 = (2, 3)
perpendicular_point2 = (8, 5)
perpendicular_point3 = (4, 8)
perpendicular_point4 = (6, 2)

plt.xticks([perpendicular_point1[0], perpendicular_point2[0], perpendicular_point3[0], perpendicular_point4[0]])
plt.yticks([perpendicular_point1[1], perpendicular_point2[1], perpendicular_point3[1], perpendicular_point4[1]])

draw_line(ax, *perpendicular_point1, *perpendicular_point2)
draw_line(ax, *perpendicular_point3, *perpendicular_point4)

print(f"Slope of the first line:\t{find_slope(*perpendicular_point1, *perpendicular_point2)}")
print(f"Slope of the second line:\t{find_slope(*perpendicular_point3, *perpendicular_point4)}")
```

### Numerical proof of the negative reciprocal rule

Let's say the slope of one line is $\alpha$ degrees. A line perpendicular to this line will have the slope of $\alpha + 90^\circ$ degrees.

```python
x = random.random()
alpha_rads = math.atan(x)
alpha_dgrs = math.degrees(alpha_rads)
perpendicular_angle_dgrs = alpha_dgrs + 90
perpendicular_angle_rads = math.radians(perpendicular_angle_dgrs)
y = math.tan(perpendicular_angle_rads)
print(f"Slope of first line: {x:.4f}")
print(f"Reciprocal of this slope: {1/x:.4f}")
print(f"Slope of perpendicular line: {y:.4f}")
```

### Angle of inclination
We talk about the angle that any line makes with the X axis. This angle $\theta$ is closely related to the slope of the line, as in:

\begin{align*}
m = tan(\theta)
\end{align*}

or

\begin{align*}
\theta = arctan(m)
\end{align*}

```python
x = np.linspace(-10, 10, 100)
y = 4 * x + 2

fig, ax = plt.subplots()
ax.plot(x, y, label="y = 4x + 2")
ax.axhline(y=0, color="red", linestyle="--")

angle_rad = np.arctan((y[1] - y[0]) / (x[1] - x[0]))
angle_deg = np.degrees(angle_rad)
ax.plot(x[50], y[50]-1, "o", markersize=10, markerfacecolor="none", markeredgecolor="green",
        label=f"Angle of Inclination: {angle_deg:.2f} degrees")


ax.set_xlabel("x")
ax.set_ylabel("y")

ax.legend()
```

### Distance formula

If we have $(x_{1}, y_{1})$ and $(x_{2}, y_{2})$, we can find the distance between the points using the Pythagorean theorem:

\begin{align*}
D^2 = (x_2-x_1)^2 + (y_2-y_1)^2
\end{align*}

\begin{align*}
D = \sqrt{(x_2-x_1)^2 + (y_2-y_1)^2}
\end{align*}

```python
fig, ax = plt.subplots()

dist_point1 = (1, 4)
dist_point2 = (7, 2)

draw_line(ax, *dist_point1, *dist_point2)
plt.xticks([dist_point1[0], dist_point2[0]])
plt.yticks([dist_point1[1], dist_point2[1]])


def distance(x1, y1, x2, y2):
    return np.sqrt((x2-x1)**2 + (y2-y1)**2)

print(f"Distance between the lines is:\t{distance(*dist_point1, *dist_point2)}")
```
